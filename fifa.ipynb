```python
col = ['Photo','Flag','Club Logo']
fifa01 = fifa.drop(columns = col, axis = 1)
fifa01.shape
```

# VALUE AND WAGE 

```python
def fun_convert(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1]
        
        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
fifa01['Value']=fifa01['Value'].apply(fun_convert)
fifa01['Wage'] = fifa01['Wage'].apply(fun_convert)
fifa01
```

# JOINED DATA CONVERSION

```python
def date_conversion(date):
    if(pd.isna(date))!= True:
        val = []
        val = date.split(',')
        dat = int(val[1])
        return dat
    else:
        return date
fifa01['Joined']=fifa01['Joined'].apply(date_conversion)
```

# HEIGHT CONVERSION

```python
def height_conversion(height):
    if(pd.isna(height))!= True:
        chk = str(height)
        h = []
        h = chk.split("'")  
        ft = float(h[0])
        if( h[1] != ''):
            inch = float(h[1])
        else:
            inch = 0
        tot_inc = inch + ft*12
        h = tot_inc * 2.54
        return h
    else:
        return height
    
fifa01['Height'] = fifa01['Height'].apply(height_conversion)
```

```python
fifa01['Height'] = fifa01['Height'].fillna(value = fifa01['Height'].mean())
```

# WEIGHT COLUMN CONVERSION

```python
def weight_conversion(weight):
    if(pd.isna(weight))!= True:
        w = int(weight[0:-3])*0.453592
        return w
    else:
        return weight
    
fifa01['Weight'] = fifa01['Weight'].apply(weight_conversion)
```

```python
fifa01['Weight'] = fifa01['Weight'].fillna(value = fifa01['Weight'].mean())
```

# RELEASE CLAUSE CONVERSION

```python
def fun_convert(df_value):
    
    if(pd.isna(df_value))!= True:
        try:
            value = float(df_value[1:-1])
            suffix = df_value[-1]

            if suffix == 'M':
                value = value * 1000000
            elif suffix == 'K':
                value = value * 1000
        except ValueError:
            value = 0
        return value
    else:
        return df_value

fifa01['Release Clause']=fifa01['Release Clause'].apply(fun_convert)
```

```python
#Choose the best squad
BestSquad_DF = fifa[['Name', 'Age', 'Overall', 'Potential', 'Position']]
BestSquad_DF.head()
```

```python
def find_best_squad(position):
    BestSquad_DF_copy = BestSquad_DF.copy()
    BestSquad = []
    for i in position:
        BestSquad.append([i,BestSquad_DF_copy.loc[[BestSquad_DF_copy[BestSquad_DF_copy['Position'] == i]['Overall'].idxmax()]]['Name'].to_string(index = False), BestSquad_DF_copy[BestSquad_DF_copy['Position'] == i]['Overall'].max()])
        BestSquad_DF_copy.drop(BestSquad_DF_copy[BestSquad_DF_copy['Position'] == i]['Overall'].idxmax(), inplace = True)

    return pd.DataFrame(np.array(BestSquad).reshape(11,3), columns = ['Position', 'Player', 'Overall']).to_string(index = False)
```

```python
squad_Formation433 = ['GK', 'LB', 'CB', 'CB', 'RB', 'LM', 'CDM', 'RM', 'LW', 'ST', 'RW']
print ('Best Squad of Formation 4-3-3')
print (find_best_squad(squad_Formation433))
```

```python
squad_Formation3412 = ['GK', 'CB', 'CB', 'CB', 'LM', 'CM', 'CM', 'RM', 'CAM', 'ST', 'ST']
print ('Best Squad of Formation 3-4-1-2')
print (find_best_squad(squad_Formation3412))
```
